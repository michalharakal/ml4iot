## Sinus

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values (from -2*pi to 2*pi)
x = np.linspace(-2 * np.pi, 2 * np.pi, 100)

# Generate corresponding y values (sine of x)
y = np.sin(x)

# Plot the sine wave
plt.plot(x, y)

# Add labels and title
plt.xlabel('$x$')
plt.ylabel('$sin(x)$')
plt.grid(True)

plt.xticks([-2 * np.pi, -np.pi, 0, np.pi, 2 * np.pi],
           ['-2π', '-π', '0', 'π', '2π'])

plt.yticks([-1, -0.5, 0, 0.5, 1],
           ['-1', '-0.5', '0', '0.5', '1'])

# Show the plot
#plt.show()
plt.savefig("sinus.svg", dpi=350)
```

## Sinus Apollo Mission

Sources:
* [Annotated sin/cos implementation](https://fermatslibrary.com/s/apollo-11-implementation-of-trigonometric-functions)

)

$\frac{1}{2}\sin\left(\frac{\pi}{2}x\right)$

Apollo Mission Aproximation
$y(x)=0.7853134 \cdot x - 0.3216147^3 + 0.036551 \cdot x^5$

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values (from 0 to pi)
x = np.linspace(0, np.pi, 100)

# Generate corresponding y values
pi_half = np.pi / 2
y = np.sin(pi_half * x) / 2
y_linear = 0.7853134 * x - 0.3216147 * (x ** 3) + 0.036551 * (x ** 5)

# Plot the sine wave
plt.plot(x, y)
plt.plot(x, y_linear, label='Linear Approximation', color="red", linestyle='--')

# Add labels and title
plt.xlabel('$x$')
plt.ylabel('$\\frac{1}{2}sin\\left(\\frac{\pi}{2}x\\right)$')
plt.grid(True)

plt.xticks([0, np.pi / 2, np.pi],
           ['0', 'π/2', 'π'])

# Show the plot
#plt.show()
plt.savefig("apollo_sinus.svg", dpi=350)
```

## ReLU

```python
import numpy as np
import matplotlib.pyplot as plt


# Define the ReLU function
def relu(x_value: float) -> float:
    return np.maximum(0, x_value)


# Generate x values
x = np.linspace(-2, 2, 100)

# Calculate corresponding y values using ReLU function
y = relu(x)

# Plot the ReLU function
plt.plot(x, y)

# Add labels and title
plt.xlabel('x')
plt.ylabel('ReLU(x)')
plt.title('ReLU Function')
plt.grid(True)



plt.savefig("relu.svg", dpi=350)
```
